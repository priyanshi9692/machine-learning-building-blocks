## Machine learning cycle

The machine learning cycle, often referred to as the "machine learning workflow" is a systematic series of steps that data scientists or machine learning engineers follow to develop and deploy machine learning models. While the specific steps and their order can vary depending on the project and domain, here are the fundamental stages of the machine learning cycle:

1. Data Collection
2. Data Exploration and Feature Engineering (aka data preprocessing)
3. Model Training/Validation
4. Model Deployment/Maintenance

### Data Collection

Data collection is a systematic process designed to obtain valuable information, enabling the creation of a comprehensive and coherent dataset tailored for a particular business objective. 
This is a crucial step for various data-driven initiatives and ventures, encompassing fields like business intelligence, machine learning, and large-scale data analytics. 

### Data Exploration and Feature Engineering

Data exploration has two primary goals: To highlight traits of single variables, and reveal patterns and relationships between variables.
Feature engineering involves data cleaning, data normalization/standardization, identifying principal component (PCA, data reduction), data balancing, data splitting (train/test/validation), etc.

Key Points:

1. The training set is used for model learning.
2. The validation set is used for hyperparameter tuning and model selection.
3. The test set is used for unbiased evaluation of the model's performance.

**Note: Training set, validation set and Test set should belong to same distribution**

### Model Training/Validation

Model training is a fundamental step in machine learning where a machine learning algorithm learns patterns and relationships within a dataset to make predictions or decisions. This process involves exposing the algorithm to labeled training data and adjusting its internal parameters to minimize the difference between its predictions and the actual outcomes.

### Model deplyment/Maintenance

So, once model is trained with a decent training and validation accuracy, we can then deploy the model, in simple terms that means now this model is ready to see the real-time data and make predictions on that.  
And model maintenance is continuous monitoring of the incoming data to detect data drift (changes in data distribution) and model drift (degradation in model performance) and updating the model if drift is detected. 


## Different types of Machine learning algorithms



1. Supervised machine learning, such as Support vector machines, Logistic regression, Decision trees, Random forest, KNN (K nearest neighbor, etc)
2. Unsupervised machine learning, such as K-means clustering
3. Semi-supervised learning, example: in many practical situations, the cost to label is quite high, since it requires skilled human experts to do that. So, in the absence of labels in the majority of the observations but present in few, semi-supervised algorithms are the best candidates for the model building.
4. Reinforcement Learning, aims at using observations gathered from the interaction with the environment to take actions that would maximize the reward or minimize the risk. Reinforcement learning algorithm (called the agent) continuously learns from the environment in an iterative fashion. In the process, the agent learns from its experiences of the environment until it explores the full range of possible states.

## Supervised Machine learning models

Broadly categorized into two main types:

1. Classification: Classification models are used when the target variable (the variable you're trying to predict) is a categorical or discrete variable. This means that it falls into one of several predefined classes or categories.


    Example Applications:
    1. Spam email detection (categorizing emails as spam or not spam)
    2. Image classification (recognizing objects in images, e.g., cats vs. dogs)
    3. Disease diagnosis (categorizing patients as having a disease or not)


2. Regression:  Regression models are used when the target variable is a continuous numerical variable. The goal is to predict or estimate this numerical value.

    Example Applications:
    1. Predicting house prices based on features like size, location, and number of bedrooms.
    2. Forecasting stock prices based on historical data and market indicators.
    3. Estimating a person's age based on demographic information.

# Jupyter Notebook

Jupyter Notebook is a notebook authoring application. It is an open source web application that allows you to create and share documents containing live code, equations, visualizations and narrative text. 

It is widely used for interactive computing and data analysis, particularly in the fields of Data Science or Machine learning.

## Installing Jupyter

To install and use Jupyter Notebook, you'll need Python installed on your system. Here are the prerequisite steps and installation instructions:

### Prerequisites:

Python: Ensure Python is installed on your system. Jupyter Notebook works with Python 3.3 or later. You can check your Python version by running python --version or python3 --version in your terminal.
Installation:

Install Python: If you don't have Python installed, download it from the official Python website: Python Downloads. Follow the installation instructions for your specific operating system.

Install pip (Python Package Manager): Most Python installations come with pip pre-installed. You can check if you have it by running pip --version or pip3 --version. If it's not installed, you can follow the instructions here: pip Installation.


### Create Python Virtual Environment

Why do we need virtual environment,
1. Isolation of Dependencies: Different projects may require different versions of packages or even different versions of Python itself. Virtual environments allow you to isolate the dependencies for each project, preventing conflicts.

2. Clean development environment: Virtual environments provide a clean slate for each project, avoiding the clutter that can occur when installing packages globally. This helps in maintaining a clean and reproducible development environment.

3. Avoiding System-wide changes: Installing packages system-wide could lead to unintended side effects, especially if you're working on a shared machine or server. Virtual environments keep the changes localized to the project.

4. Dependency Management: With virtual environments, you can install and manage project-specific dependencies without affecting the global Python installation. This is particularly useful when working on projects with conflicting requirements.

Open your terminal, and Use following commands,

`python3 -m venv venv`

Next source your virtual environment,

`source venv/bin/activate`

#### Jupyter notebook shortcuts

In Jupyter Notebooks, there are several shortcuts that can help you quickly access Python methods and documentation. Here are some handy shortcuts:

1. Autocomplete / Intellisense: Press Tab for autocompletion after typing the first few letters of a variable or method name.
    
2. Function Signature: After typing a function name, press Shift + Tab inside the parentheses to see the function signature and docstring.

3. Cell Execution: Press Shift + return (Enter) to run the current cell and move to the next one.

4. Inserting a New Cell: Press **B** to insert a new cell **below** the current one. Press **A** to insert a new cell **above** the current one.

5. Deleting a Cell: Press **D** twice to delete the current cell.

6. Changing Cell Type:
    
    A. Press Y to change the cell type to Code. 
    B. Press M to change the cell type to Markdown.

7. Save Notebook: Press S to save the notebook.


### Exiting Jupyter Notebook:

1. To exit Jupyter Notebook, save your work, close the tabs, and return to your terminal. You can stop the Jupyter Notebook server by pressing Ctrl+C in the terminal.

**Remember to save your work regularly by clicking the "Save" button or using the keyboard shortcut (Ctrl+S on windows or Cmd+S on macOS).**
