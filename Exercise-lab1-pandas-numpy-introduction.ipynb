## Plots

### Age and Sex

Let us plot a some charts to get more insights of the dataset, and understand what are the features impacting a person's survival in Titanic

```python
# Creating labels for figures
survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 8))

# Extract exclusive rows where Sex is Female and Male
women = train_df[train_df['Sex']=='female']
men = train_df[train_df['Sex']=='male']

# Plot histplot for the Survived and Not survives age groups for Female
ax = sns.histplot(women[women['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[0], kde =False)
ax = sns.histplot(women[women['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[0], kde =False)
ax.legend()
ax.set_title('Female')


# Plot histplot for the Survived and Not survives age groups for Female
ax = sns.histplot(men[men['Survived']==1].Age.dropna(), bins=18, label = survived, ax = axes[1], kde = False)
ax = sns.histplot(men[men['Survived']==0].Age.dropna(), bins=40, label = not_survived, ax = axes[1], kde = False)
ax.legend()
_ = ax.set_title('Male')
```

The data from the Titanic reveals an interesting pattern: 

1. Men between the ages of 18 and 30 had the best chances of survival though the same age group has more number of men who did not survive. This age group shows the highest probability of making it through the disaster. However, this trend is not limited to men; it also applies to women, especially those between the ages of 14 and 40.

2. For men, there was a significant drop in survival probability for those aged between 5 and 18. Surprisingly, this dip is not observed among women. Additionally, infants had a slightly better chance of survival.

3. Because specific age ranges seem to correlate with increased odds of survival, and to ensure that all the features we analyze are on a similar scale, we will create group of passengers based on the age categories later in our analysis.

### Embarked, Pclass and Sex

FacetGrid chart: This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called ‚Äúlattice‚Äù, ‚Äútrellis‚Äù, or ‚Äúsmall-multiple‚Äù graphics.

Seaborn's pointplot function to visualize the relationships between the 'Pclass' (passenger class), 'Survived' (survival status), and 'Sex' (gender) variables. It likely uses different colors to distinguish between categories within the 'Sex' variable. The FacetGrid organizes these plots into a grid, which can be useful for comparing different subsets of the data

```python
FacetGrid = sns.FacetGrid(train_df, row='Embarked', height=4, aspect=1.8)
FacetGrid.map(sns.pointplot, 'Pclass', 'Survived', 'Sex', palette='dark:#1f77b4', order=None, hue_order=None )
FacetGrid.add_legend()
```

Embarked seems to be correlated with survival, depending on the gender.

Women on port Q had a higher chances of survival. Men have a high survival probability if they were on port S and C, but a low probability if they are on port Q.

Pclass also seems to be correlated with survival. We will generate another plot of it below.

### SibSp and Parch

SibSp and Parch would make more sense as a combined feature, that shows the total number of relatives, a person has on the Titanic. Let's create a plot create that can help us to see how many passengers were not alone.

```python
data = [train_df, test_df]
for dataset in data:
    dataset['relatives'] = dataset['SibSp'] + dataset['Parch']
    dataset.loc[dataset['relatives'] > 0, 'not_alone'] = 0
    dataset.loc[dataset['relatives'] == 0, 'not_alone'] = 1
    dataset['not_alone'] = dataset['not_alone'].astype(int)
train_df['not_alone'].value_counts()
```
